```python
import keras
from keras.datasets import mnist
from keras.models import Sequential
from keras.layers import Dense, Dropout
from keras.optimizers import RMSprop
```

```python
from sklearn.model_selection import GridSearchCV
from keras.wrappers.scikit_learn import KerasClassifier
```

```python
from keras.datasets import mnist
(x_train, y_train), (x_test, y_test) = mnist.load_data()
```

```python
#batch_size = 128
#epochs = 50
num_classes = 10
```

```python
x_train = x_train.reshape(60000, 784)
x_test = x_test.reshape(10000, 784)
```

```python
x_train = x_train.astype('float32')
x_test = x_test.astype('float32')
x_train /= 255
x_test /= 255
```

```python
y_train = keras.utils.to_categorical(y_train, num_classes)
y_test = keras.utils.to_categorical(y_test, num_classes)
```

```python
def make_model(activation='relu', out_dim=512, dropout=0.2): 
    model  = Sequential()
    model.add(Dense(out_dim, activation = activation, input_shape = (784,)))
    model.add(Dropout(dropout))
    model.add(Dense(out_dim, activation = activation))
    model.add(Dropout(dropout))
    model.add(Dense(10, activation='softmax'))
    model.summary()
    model.compile(loss='categorical_crossentropy', optimizer=RMSprop(lr=0.0009), metrics=['accuracy'])
    make_model = model
    return make_model
```

```python
activation = ["relu", "sigmoid"]
out_dim = [512]
dropout = [0.2]
batch_size = [128, 256]
epochs = [110]
```

```python
estimator = KerasClassifier(build_fn = make_model, verbose = 1)
param_grid = dict(activation = activation
                 ,out_dim = out_dim
                 ,dropout = dropout
                 ,batch_size = batch_size
                 ,epochs = epochs)
model = GridSearchCV(estimator = estimator, param_grid = param_grid )
```
